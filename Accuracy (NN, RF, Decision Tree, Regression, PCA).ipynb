```python
# import the required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Data reading
df = pd.read_csv("Diabetes Dataset.csv")
df.head()
```

```python
feature_cols = df.columns[0:8] # Column names of independent 
print(feature_cols)

X = df[feature_cols]  # Independent Varuable
Y = df.Outcome # Dependent Variable
print(X)
print(Y)
```

```python
# 25% testing, 75% training
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

### Random Forest

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

# Create a Gaussian Classifier
RF=RandomForestClassifier(n_estimators=100,random_state = 1)

# Train the model
RF.fit(X_train,Y_train)

# Run the prediction using testing set
Y_pred = RF.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print("Accuracy of RF:",metrics.accuracy_score(Y_test, Y_pred))
```

### Linear Regression

```python
# Import Linear Regression Model
from sklearn.linear_model import LinearRegression

# Create a  Classifier
LiR = LinearRegression()

# Train the model
LiR.fit(X_train,Y_train)

# Run the prediction using testing set
Y_pred = LiR.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print("Accuracy of Linear Regression:",metrics.accuracy_score(Y_test, Y_pred))
```

### Decision Tree

```python
# Import Decision Tree Model
from sklearn import tree                                                        

# Create a  Classifier
Tree = tree.DecisionTreeClassifier()                                             

#Train the model
Tree.fit(X_train,Y_train)

# Run the prediction using testing set
Y_pred = Tree.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print("Accuracy of Decision Tree:",metrics.accuracy_score(Y_test, Y_pred))
```

### PCA

```python
X_PCA = df.drop('Outcome', 1)  
y_PCA = df['Outcome']

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(X_PCA, y_PCA, test_size=0.25, random_state=0) 

# ??
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()  
X_train_PCA = sc.fit_transform(X_train_PCA)  
X_test_PCA = sc.transform(X_test_PCA) 

# Import PCA model
from sklearn.decomposition import PCA
pca = PCA(n_components=1)  
X_train_PCA = pca.fit_transform(X_train_PCA)  
X_test_PCA = pca.transform(X_test_PCA) 

# Use random forest for making prediction
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(max_depth=2, random_state=0)  
classifier.fit(X_train_PCA, y_train_PCA)

# Predicting the Test set results
y_pred_PCA = classifier.predict(X_test_PCA) 

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print("Accuracy of PCA-RF(n=1):",metrics.accuracy_score(y_test_PCA, y_pred_PCA))
```

```python
from sklearn.neural_network import MLPClassifier
NN = MLPClassifier(solver = 'lbfgs',alpha=1e-5, hidden_layer_sizes = (15,),random_state=1)

NN.fit(X_train,Y_train)

Y_pred = NN.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
print("Accuracy of Neural Network:",metrics.accuracy_score(Y_test, Y_pred))
```
